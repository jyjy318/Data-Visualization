```python
# 세종시 클러스터
import numpy as np
import pandas as pd
import json
import requests
import folium
data_jong = pd.read_csv('./data/세종시 세대별 갭투자 현황.csv')
data_jong

# state_geo = './data/33.세종시_지적도.geojson'
# state_geo2 = json.load(open(state_geo, encoding = 'utf-8'))

from folium import plugins
# 지도를 불러들이기
map_jong = folium.Map(location=['36.5', '127.3'], zoom_start='11')
regional_count = data_jong[['위도', '경도']]

# 지도에 색 적용 및 데이터 연결

#지도 전체화면 추가 코드
plugins.Fullscreen(position = 'topright',
                  title = "Click to Expand",
                  title_cancel = "Click to Exit",
                  force_separate_button = True).add_to(map_jong)


plugins.MousePosition().add_to(map_jong)
plugins.MarkerCluster(regional_count).add_to(map_jong)
map_jong.save('세종 갭투자 ver2.html')
map_jong
```

```python
import plotly.graph_objects as go

data = [1, 2, 3, 4, 5, 6, 7, 8, 9]

fig = go.Figure()
fig.add_trace(go.Box(y=data, quartilemethod="linear", name="Linear Quartile Mode"))
fig.add_trace(go.Box(y=data, quartilemethod="inclusive", name="Inclusive Quartile Mode"))
fig.add_trace(go.Box(y=data, quartilemethod="exclusive", name="Exclusive Quartile Mode"))
fig.update_traces(boxpoints='all', jitter=0)
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

data_seoul = pd.read_csv('./data/서울시 세대별 갭투자 현황.csv')
data_jong = pd.read_csv('./data/세종시 세대별 갭투자 현황.csv')

y0 = data_seoul['갭']
y1 = data_jong['갭']

fig = go.Figure()
fig.add_trace(go.Box(y=y0, name='서울시 세대별 갭',
                marker_color = 'indianred'))
fig.add_trace(go.Box(y=y1, name = '세종시 세대별 갭',
                marker_color = 'lightseagreen'))

fig.show()
```
